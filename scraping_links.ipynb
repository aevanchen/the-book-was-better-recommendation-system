```python
from tqdm import tqdm
```

```python
movie_dict={}

for movie_name in tqdm(movies):
    arguments = {"keywords":"{}(Moive)".format(movie_name),"limit":1,"print_urls":True,"no_download":True}
    link=response.download(arguments)   #passing the arguments to the function#creating list of arg
    movie_dict[movie_name]=link
    
```

```python
import regex as re
re.sub(r'[^\w\s]','',movies[4202])
```

```python
for movie_name in tqdm(movies[4202:]):
    
    try:
        arguments = {"keywords":"{}(Moive)".format(movie_name),"limit":1,"print_urls":True,"no_download":True}
        link=response.download(arguments) 
    except:
        arguments = {"keywords":"{}(Moive)".format(re.sub(r'[^\w\s]','',movie_name)),"limit":1,"print_urls":True,"no_download":True}
        link=response.download(arguments) 
    
    movie_dict[movie_name]=link
```

```python
book_dict={}

for book_name in tqdm(books):
    try:
        arguments = {"keywords":"{} (book)".format(book_name),"limit":1,"print_urls":True,"no_download":True}
        link=response.download(arguments)   #passing the arguments to the function#creating list of arg
    except:
        arguments = {"keywords":"{}(book)".format(re.sub(r'[^\w\s]','',book_name)),"limit":1,"print_urls":True,"no_download":True}
        link=response.download(arguments) 
    
    book_dict[book_name]=link
```
